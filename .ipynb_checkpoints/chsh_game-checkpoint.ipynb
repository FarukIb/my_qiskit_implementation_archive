```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, array_to_latex
from qiskit.result import marginal_distribution
from qiskit.circuit.library import UGate
from math import pi
import random
```

```python
alpha = 0
beta = pi / 8.0
quarter = pi / 4.0

def make_circuit(p, q):
    circuit = QuantumCircuit(2)

    # put the qbit in bell state, phi+
    circuit.h(0)
    circuit.cx(0, 1)

    circuit.barrier()

    # rotate depending on question
    if (p == 0):
        circuit.ry(alpha * 2, 0)
    else:
        circuit.ry((alpha + quarter) * 2, 0)

    if (q == 0):
        circuit.ry(beta * 2, 1)
    else:
        circuit.ry((beta - quarter) * 2, 1)

    circuit.barrier()

    # measure
    circuit.measure_all()
    return circuit
```

```python
display(make_circuit(0,0).draw(output="mpl"))
```

```python
def strategy(x, y):
    result = AerSimulator().run(make_circuit(x,y), shots=1).result()
    statistics = result.get_counts()

    bits = list(statistics.keys())[0]
    a, b = bits[0], bits[1]
    return a, b
```

```python
def play_game(strategy):
    x, y = random.randint(0, 1), random.randint(0, 1)

    a, b = strategy(x, y)

    if ((a != b) == (x & y)):
        return 1
    return 0
```

```python
NUM_GAMES = 1000
TOT_SCORE = 0
for i in range(0, NUM_GAMES):
    TOT_SCORE += play_game(strategy)
print(float(TOT_SCORE) / float(NUM_GAMES))
```
